```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/RuthieRB/Testrepository/refs/heads/main/LB1/CODE/titanic.csv'
df = pd.read_csv(url)
df.head()
```

```python
sns.displot(df['Sex'])
```

```python
sns.displot(df['Age'], kde=True, bins=20)
```

```python
sns.jointplot(x='Age', y='Fare', data=df, kind='hex')
```

```python
sns.pairplot(df, hue='Sex', palette='Blues')
```

```python
sns.rugplot(df['Age'])
```

```python
sns.set_style('dark')
plt.figure(figsize=(10,7))
sns.set_context('poster', font_scale=0.7)
sns.countplot(x='Sex', data=df)
```

```python
sns.barplot(x='Sex', y='Fare', data=df)
```

```python
sns.countplot(x='Sex', data=df)
```

```python
sns.boxplot(x='Fare', y='Age', data=df, hue='Sex')
plt.legend(loc=0)
```

```python
sns.violinplot(x='Fare', y='Age', data=df, hue='Sex')
```

```python
plt.figure(figsize=(10,7))
sns.stripplot(x='Fare', y='Age', data=df, jitter=True, hue='Sex', palette='summer')
sns.set_style('darkgrid')
```

```python
sns.violinplot(x='Fare', y='Age', data=df, hue='Sex')
```
